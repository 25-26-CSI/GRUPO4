# Cifrado por sustitución simple con diccionario

# Código

```python
# Diccionario de sustitución
# Aquí definimos qué símbolo reemplaza a cada letra del abecedario y el espacio.
# Clave = letra original, Valor = símbolo que la sustituye
tabla_sustitucion = {
    'a': 'ღ', 'b': '☢', 'c': 'a', 'd': '✘', 'e': '☞',
    'f': '⊗', 'g': '⚔', 'h': '♣', 'i': '✦', 'j': '∑',
    'k': '₪', 'l': '∟', 'm': '⚙', 'n': 'Ω', 'o': '¡',
    'p': '☭', 'q': '§', 'r': '¥', 's': '☯', 't': '⚡',
    'u': '❤', 'v': '✧', 'w': '⇆', 'x': 'þ', 'y': '∪',
    'z': '♛', ' ': '▣'
}

# Crear diccionario inverso para descifrar
# Esto genera un nuevo diccionario donde la clave es el símbolo y el valor la letra original
# Se usa para poder revertir el cifrado y recuperar el texto original
tabla_descifrado = {v: k for k, v in tabla_sustitucion.items()}

# Función para cifrar texto
def cifrar(texto):
    # Recorre cada carácter del texto
    # Si el carácter está en la tabla de sustitución, lo reemplaza
    # Si no, lo deja igual (por ejemplo números o signos de puntuación)
    # ''.join(...) une todos los caracteres resultantes en un solo string
    return ''.join([tabla_sustitucion.get(c.lower(), c) for c in texto])

# Función para descifrar texto
def descifrar(texto):
    # Recorre cada carácter del texto cifrado
    # Si el carácter está en la tabla inversa, lo reemplaza por la letra original
    # Si no, lo deja igual
    return ''.join([tabla_descifrado.get(c, c) for c in texto])

# Función para mostrar el abecedario con números y sustitutos
def mostrar_mapa_numerado(tabla):
    # Creamos un string con todas las letras mayúsculas del abecedario
    abecedario_normal = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    # Encabezado de la tabla
    print("Num | Letra | Sustituto")
    print("------------------------")

    # Recorremos cada letra y su número correspondiente (1 a 26)
    for i, letra in enumerate(abecedario_normal, start=1):
        # Obtenemos el símbolo correspondiente del diccionario
        # Si no está (por si acaso), usamos la misma letra
        sustituto = tabla.get(letra.lower(), letra)
        # Imprimimos el número, la letra y su símbolo de sustitución
        print(f"{i:2}  |   {letra}   |    {sustituto}")
# Mostramos el mapa numerado de todas las letras y sus símbolos
print("\nMapa de sustitución numerado:")
mostrar_mapa_numerado(tabla_sustitucion)
```

# Parámetros de uso

```python
mensaje = "ATAQUEINMINENTE"
```

```python
mensaje = "UNIVERSIDAD CENTRAL DEL ECUADOR"
```

# Cifrado

```python
cifrado = cifrar(mensaje)
print("Cifrado:", cifrado)
```

# Descifrado

```python
descifrado = descifrar(cifrado)
print("Descifrado:", descifrado)
```
